```python
# Q4. Describe the process of feature engineering in the context of the student performance data set. How
# did you select and transform the variables for your model?
import pandas as pd
df=pd.read_csv("stud (1).csv")
```

```python
"""gender is a catagorical feature having only two values ,male and female , female is coded 0
and male is coded 1 so we will use here dictionary"""
df["gender"] = df["gender"].replace({"female":0, "male":1})
```

```python
df
```

```python
"""since  test_preparation_course has only two values then we will use dictionary to convert it to numerical"""
df["test_preparation_course"] = df["test_preparation_course"].replace({"completed":0, "none":1})
```

```python
df
```

```python
"""since  lunch has only two values then we will use dictionary to convert it to numerical"""
df["lunch"] = df["lunch"].replace({"standard":0, "free/reduced":1})
```

```python
df
```

```python
"""we will use one hot encoder for race_ethnicity and ordinal encoder for parental_level_of_education """
from sklearn.preprocessing import OneHotEncoder,OrdinalEncoder
import pandas as pd

encoder = OneHotEncoder()
encoded = encoder.fit_transform(df[["race_ethnicity"]])

columns = encoder.get_feature_names_out()
df1 = pd.DataFrame(encoded.toarray(), columns=columns)
df2=pd.concat([df,df1],axis=1)
```

```python
df2=df2.drop(["race_ethnicity"],axis=1)
```

```python
df2
```

```python


from sklearn.preprocessing import OrdinalEncoder


encoder = OrdinalEncoder(categories=[["some college","high school","master's degree"," associate's degree","bachelor's degree",'some high school', "associate's degree"]])


encoded_data = encoder.fit_transform(df[["parental_level_of_education"]])

df_encoded = pd.DataFrame(encoded_data,columns=["parental education"]) 


```

```python
df_final=pd.concat([df2,df_encoded],axis=1)
df_final.head()
```

```python
"""In the above dataset "residual sugar","chlorides","total sulfur dioxide","sulphates" do not show normal behaviour
so we apply log transformation to get the normalised output  """



import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt


df = pd.read_csv("winequality-red.csv", sep=";")
for i in ["residual sugar","chlorides","total sulfur dioxide","sulphates"]:
    df[i] = np.log(df[i])
    sns.histplot(data=df, x=i,kde=True)
    plt.show()
```

```python
# Q6. Using the wine quality data set, perform principal component analysis (PCA) to reduce the number of
# features. What is the minimum number of principal components required to explain 90% of the variance in
# the data?

import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

wine_data = pd.read_csv("winequality-red.csv",sep=";")

scaler = StandardScaler()
scaled_data = scaler.fit_transform(wine_data)
print(scaled_data)

pca = PCA()
pca=pca.fit(scaled_data)


variance = pca.explained_variance_ratio_
print(variance)
cumulative_variance = np.cumsum(variance)
num_components = len(cumulative_variance[cumulative_variance <= 0.9])

print(f"The minimum number of principal components required to explain 90% of the variance in the data is {num_components}")
```
